```python
# Pivot table view: Week vs Horizon for each LG
import pandas as pd

for lg in ['TRR', 'TRP']:
    print(f"\n{'='*60}")
    print(f"{lg}: ML WAPE by Week x Horizon")
    print("="*60)
    lg_data = metrics_lg[metrics_lg['liquidity_group'] == lg]
    
    # ML WAPE pivot
    ml_pivot = lg_data.pivot(index='week_start', columns='horizon', values='ml_wape')
    print("ML WAPE:")
    print(ml_pivot.round(4))
    
    # LP WAPE pivot  
    print(f"\n{lg}: LP WAPE by Week x Horizon")
    lp_pivot = lg_data.pivot(index='week_start', columns='horizon', values='lp_wape')
    print("LP WAPE:")
    print(lp_pivot.round(4))
    
    # Difference (negative = ML better)
    print(f"\n{lg}: ML - LP WAPE (negative = ML better)")
    diff_pivot = ml_pivot - lp_pivot
    print(diff_pivot.round(4))
```

```python
# Aggregate metrics by week and LG (averaging across horizons)
# Or if you want per-horizon, we can adjust

# Option 1: Average WAPE across all horizons per (week, LG)
summary_by_week_lg = metrics_lg.groupby(['week_start', 'liquidity_group']).agg({
    'ml_wape': 'mean',
    'lp_wape': 'mean',
    'n_obs': 'sum'
}).reset_index()

# TRR Summary
print("=" * 60)
print("TRR: Weekly ML vs LP WAPE")
print("=" * 60)
trr_summary = summary_by_week_lg[summary_by_week_lg['liquidity_group'] == 'TRR'][['week_start', 'ml_wape', 'lp_wape', 'n_obs']]
trr_summary['ml_better'] = trr_summary['ml_wape'] < trr_summary['lp_wape']
print(trr_summary.to_string(index=False))
print(f"\nML wins: {trr_summary['ml_better'].sum()} / {len(trr_summary)} weeks")
print(f"Avg ML WAPE: {trr_summary['ml_wape'].mean():.4f}")
print(f"Avg LP WAPE: {trr_summary['lp_wape'].mean():.4f}")

```

```python
# TRP Summary
print("\n" + "=" * 60)
print("TRP: Weekly ML vs LP WAPE")
print("=" * 60)
trp_summary = summary_by_week_lg[summary_by_week_lg['liquidity_group'] == 'TRP'][['week_start', 'ml_wape', 'lp_wape', 'n_obs']]
trp_summary['ml_better'] = trp_summary['ml_wape'] < trp_summary['lp_wape']
print(trp_summary.to_string(index=False))
print(f"\nML wins: {trp_summary['ml_better'].sum()} / {len(trp_summary)} weeks")
print(f"Avg ML WAPE: {trp_summary['ml_wape'].mean():.4f}")
print(f"Avg LP WAPE: {trp_summary['lp_wape'].mean():.4f}")
```

```python
# Filter to Tier-1 only (no passthroughs)
tier1_only = backtest_df[backtest_df['is_pass_through'] == False].copy()

print(f"Total rows: {len(backtest_df)}")
print(f"Tier-1 rows (ML predictions): {len(tier1_only)}")
print(f"Tier-2 rows (LP passthrough): {len(backtest_df) - len(tier1_only)}")

# Only h<=4 has LP baseline for comparison
tier1_h4 = tier1_only[tier1_only['horizon'] <= 4].copy()

print(f"\nTier-1 rows with LP baseline (h<=4): {len(tier1_h4)}")
```

```python
# Compute CLEAN metrics for Tier-1 only
def compute_clean_wape(df, pred_col, actual_col='actual_value'):
    """Compute WAPE for non-null predictions"""
    mask = df[pred_col].notna() & df[actual_col].notna()
    if mask.sum() == 0:
        return float('nan')
    return (df.loc[mask, actual_col] - df.loc[mask, pred_col]).abs().sum() / df.loc[mask, actual_col].abs().sum()

# Clean metrics by (week, LG) - Tier-1 only, h<=4
clean_summary = []
for (week, lg), group in tier1_h4.groupby(['week_start', 'liquidity_group']):
    clean_summary.append({
        'week_start': week,
        'liquidity_group': lg,
        'n_obs': len(group),
        'ml_wape': compute_clean_wape(group, 'y_pred_point'),
        'lp_wape': compute_clean_wape(group, 'lp_baseline_point'),
    })

clean_df = pd.DataFrame(clean_summary)
clean_df['ml_better'] = clean_df['ml_wape'] < clean_df['lp_wape']

```

```python
# TRR Clean Summary
print("=" * 70)
print("TRR: CLEAN ML vs LP WAPE (Tier-1 only, h<=4)")
print("=" * 70)
trr_clean = clean_df[clean_df['liquidity_group'] == 'TRR']
print(trr_clean.to_string(index=False))
print(f"\nML wins: {trr_clean['ml_better'].sum()} / {len(trr_clean)} weeks")
print(f"Avg ML WAPE: {trr_clean['ml_wape'].mean():.4f} ({trr_clean['ml_wape'].mean()*100:.1f}%)")
print(f"Avg LP WAPE: {trr_clean['lp_wape'].mean():.4f} ({trr_clean['lp_wape'].mean()*100:.1f}%)")

```

```python
# TRP Clean Summary  
print("\n" + "=" * 70)
print("TRP: CLEAN ML vs LP WAPE (Tier-1 only, h<=4)")
print("=" * 70)
trp_clean = clean_df[clean_df['liquidity_group'] == 'TRP']
print(trp_clean.to_string(index=False))
print(f"\nML wins: {trp_clean['ml_better'].sum()} / {len(trp_clean)} weeks")
print(f"Avg ML WAPE: {trp_clean['ml_wape'].mean():.4f} ({trp_clean['ml_wape'].mean()*100:.1f}%)")
print(f"Avg LP WAPE: {trp_clean['lp_wape'].mean():.4f} ({trp_clean['lp_wape'].mean()*100:.1f}%)")

```
