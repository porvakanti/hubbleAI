```python
# Compute CLEAN metrics for Tier-1 only
def compute_clean_wape(df, pred_col, actual_col='actual_value'):
    """Compute WAPE for non-null predictions"""
    mask = df[pred_col].notna() & df[actual_col].notna()
    if mask.sum() == 0:
        return float('nan')
    return (df.loc[mask, actual_col] - df.loc[mask, pred_col]).abs().sum() / df.loc[mask, actual_col].abs().sum()

# Clean metrics by (week, LG) - Tier-1 only, h<=4
clean_summary = []
for (week, lg), group in tier1_h4.groupby(['week_start', 'liquidity_group']):
    clean_summary.append({
        'week_start': week,
        'liquidity_group': lg,
        'n_obs': len(group),
        'ml_wape': compute_clean_wape(group, 'y_pred_point'),
        'lp_wape': compute_clean_wape(group, 'lp_baseline_point'),
    })

clean_df = pd.DataFrame(clean_summary)
clean_df['ml_better'] = clean_df['ml_wape'] < clean_df['lp_wape']

```

```python
import pandas as pd
import numpy as np

# Load backtest predictions
backtest = pd.read_parquet(status.output_paths["backtest"])

# Load alpha table for reference
alpha_df = pd.read_parquet(status.metrics_paths["alpha_by_lg_horizon"])
print("=== Alpha Table (TRP only) ===")
print(alpha_df[alpha_df.liquidity_group == "TRP"][["liquidity_group", "horizon", "alpha", "wape_ml", "wape_lp", "wape_hybrid"]])
print()

# Filter to Tier-1 only (true ML predictions)
tier1 = backtest[backtest["is_pass_through"] == False].copy()

```

```python
# Aggregate-then-error WAPE function
def wape_agg(actual, pred):
    """Treasury-style WAPE: |sum(actual) - sum(pred)| / |sum(actual)|"""
    total_actual = actual.sum()
    total_pred = pred.sum()
    if abs(total_actual) < 1e-9:
        return np.nan
    return abs(total_actual - total_pred) / abs(total_actual)

# Compute per-week WAPE for each LG and horizon
def compute_weekly_wape(df, lg, horizon):
    """Compute per-week WAPE for ML, LP, Hybrid for a specific LG and horizon."""
    subset = df[(df["liquidity_group"] == lg) & (df["horizon"] == horizon)].copy()
    
    if subset.empty:
        return pd.DataFrame()
    
    results = []
    for week, group in subset.groupby("week_start"):
        actual = group["actual_value"]
        ml = group["y_pred_point"]
        lp = group["lp_baseline_point"]
        hybrid = group["y_pred_hybrid"]
        
        # Skip if no valid data
        if actual.isna().all():
            continue
            
        result = {
            "week_start": week,
            "liquidity_group": lg,
            "horizon": horizon,
            "n_entities": len(group),
            "ml_wape": wape_agg(actual, ml),
            "lp_wape": wape_agg(actual.dropna(), lp.dropna()) if lp.notna().any() else np.nan,
            "hybrid_wape": wape_agg(actual, hybrid),
        }
        
        # Add win indicators
        if not np.isnan(result["lp_wape"]):
            result["ml_beats_lp"] = result["ml_wape"] < result["lp_wape"]
            result["hybrid_beats_lp"] = result["hybrid_wape"] < result["lp_wape"]
            result["hybrid_beats_ml"] = result["hybrid_wape"] < result["ml_wape"]
        else:
            result["ml_beats_lp"] = np.nan
            result["hybrid_beats_lp"] = np.nan
            result["hybrid_beats_ml"] = np.nan
            
        results.append(result)
    
    return pd.DataFrame(results)

```

```python
# Compute for TRP horizons 1-4 (where hybrid matters)
print("=== Per-Week WAPE Comparison (Test Set) ===")
print()

for horizon in [1, 2, 3, 4]:
    weekly = compute_weekly_wape(tier1, "TRP", horizon)
    
    if weekly.empty:
        print(f"TRP H{horizon}: No data")
        continue
    
    # Get alpha for this LG×horizon
    alpha_row = alpha_df[(alpha_df["liquidity_group"] == "TRP") & (alpha_df["horizon"] == horizon)]
    alpha_val = alpha_row["alpha"].values[0] if len(alpha_row) > 0 else 1.0
    
    print(f"=== TRP Horizon {horizon} (α = {alpha_val:.1f}) ===")
    print(weekly[["week_start", "ml_wape", "lp_wape", "hybrid_wape", "hybrid_beats_lp"]].to_string(index=False))
    
    # Summary stats
    print(f"\nSummary:")
    print(f"  ML wins vs LP:     {weekly['ml_beats_lp'].sum():.0f} / {weekly['ml_beats_lp'].notna().sum():.0f} weeks")
    print(f"  Hybrid wins vs LP: {weekly['hybrid_beats_lp'].sum():.0f} / {weekly['hybrid_beats_lp'].notna().sum():.0f} weeks")
    print(f"  Hybrid wins vs ML: {weekly['hybrid_beats_ml'].sum():.0f} / {weekly['hybrid_beats_ml'].notna().sum():.0f} weeks")
    print(f"  Mean ML WAPE:      {weekly['ml_wape'].mean():.4f}")
    print(f"  Mean LP WAPE:      {weekly['lp_wape'].mean():.4f}")
    print(f"  Mean Hybrid WAPE:  {weekly['hybrid_wape'].mean():.4f}")
    print()


```
